```python
from pandas import read_csv

pd = read_csv('../data/dm250.0_nfrb500_1024_sec_20181211-1000_1024_3sigma.amber.trigger', sep=' ')
# print(pd)
sample_vectors = np.asarray(pd[['time', 'DM']])
```
